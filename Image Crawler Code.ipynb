```python
from icrawler.builtin import BingImageCrawler
import os

# Define paths for storing images
base_dir = 'F:/Ubuntu1204/1/2/ID/'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

# Define the categories (apple and non-apple)
categories = ['apple fruit', 'oranges']

# Function to download images for a given category
def download_images(category, folder, num_images=50):
    save_dir = os.path.join(folder, category)
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    # Initialize BingImageCrawler
    crawler = BingImageCrawler(storage={'root_dir': save_dir})
    
    # Download images from Bing
    crawler.crawl(keyword=category, max_num=num_images)
    print(f"Downloaded {num_images} {category} images into {save_dir}")

# Download apple and non-apple images into train, validation, and test folders
for category in categories:
    # Download images for training set
    download_images(category, train_dir, num_images=200)
    
    # Download images for validation set
    download_images(category, validation_dir, num_images=90)
    
    # Download images for testing set
    download_images(category, test_dir, num_images=90)

print("Image download completed!")
```
